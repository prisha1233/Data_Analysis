```python
## updating above not available and unkonwn values with null
df = pd.read_csv('orders.csv',na_values = [ 'Not Available', 'unknown'])
df['Ship Mode'].unique()
#df.head(20)
```

```python
## rename columns names
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.head(5)
```

```python
### converting order date into mm-dd-yyyy
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.dtypes
```

```python
## derive new column discount, sale price and profit
## discount_amt  = list_price* discount_percent/100
## sale price = list_price - discount_price
#profit = sale_price - cost_price
df['discount_amt']  = df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price']- df['discount_amt']
df['profit'] = df['sale_price'] - df['cost_price']
df.head(10)
```

```python
## Dropping list_price, discount_percent, cost_price columns from dataset orders
df.drop(columns = ['list_price', 'discount_percent', 'cost_price'], inplace= True)
df
```
